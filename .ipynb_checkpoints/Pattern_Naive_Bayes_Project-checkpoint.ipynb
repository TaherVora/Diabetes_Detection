```python
import math 
import random 
import pandas as pd
import numpy as np


'''convert_dict = {'Pregnancies': float,'Glucose': float,'BloodPressure': float,'SkinThickness': float,'Insulin': float,'Age': float,'Outcome':np.int64}
    fileData = fileData.astype(convert_dict)'''
    
'''  method to load data from csv file''' 
def loadFile(fileName): 
    fileData = data=pd.read_csv(open(r'C:\Users\Upasana\Desktop\Assignments\PR\Second Project\diabetes.csv')) 
    return fileData 

''' method to convert all values to float as data is in text format'''     
def convertDatatoFloat(fileData): 
    dataObj = list(fileData) 
    for i in range(len(dataObj)): 
        dataRow = dataObj[i] 
        for j in range(len(dataRow)): 
            value=float(dataRow[j]) 
            dataRow[j] = value 
        dataObj[i]=dataRow 
    print("FileData in float as a list object --- ",dataObj) 
    return dataObj 

''' split dataset on the basis of third value or can say class like 0 or 1 ''' 
def prepareTrainingData(dataObj,ratio): 
    tDataSize= int(len(dataObj) * ratio) 
    tData = [] 
    testData=dataObj 
    for x in range(tDataSize): 
        position = random.randrange(len(dataObj)) 
        tData.append(testData.pop(position)) 
        print("training data for ratio 0.65 --- ",tData) 
        print("remaining data i.e test data in initial object according to ratio 0.27 --- ",testData) 
    return [tData,testData] 

''' to calculate mean and standard deviation need to categorize data according to the class''' 
def categorizeClassData(dataObj): 
    dataCategory={} 
    for i in range(len(dataObj)): 
        dataRow = dataObj[i] 
        rowLength=len(dataRow)-1 
        if(dataRow[rowLength] not in dataCategory): 
            dataCategory[dataRow[rowLength]]=[] 
        dataCategory[dataRow[rowLength]].append(dataRow) 
    print("data categorized on the basis of the class --- ",dataCategory) 
    return dataCategory 

''' method to calculate standard mean of numbers''' 
def getMean(values): 
    mean=sum(values)/float(len(values)) 
    return mean 

''' method to calculate standard deviation of all numbers'''  
def getDeviation(values): 
    average=getMean(values) 
    totalNumerate=0.0 
    for i in range(len(values)): 
        indTerm = pow((values[i]-average),2) 
        totalNumerate+=indTerm 
    variance=totalNumerate/float(len(values)) 
    deviation=math.sqrt(variance) 
    return deviation 
  
''' method to collect mean and standard deviation for each row of dataObj in a single list''' 
def prepareMeanDevList(dataObj): 
    meanDevList=[] 
    for row in zip(*dataObj): 
        dataMean = getMean(row) 
        dataDeviation = getDeviation(row) 
        meanDevList.append((dataMean,dataDeviation)) 
    ''' to remove last occurance of mean and standard deviation because of existence of class factor''' 
    meanDevList.pop() 
    print("Mean and Deviation list of all the attributes present in a dataRow except the last class name data --- ",meanDevList) 
    return meanDevList 

''' method to merge all mean and standard deviation on class basis''' 
def makeClassListMeanDev(dataObj): 
    dataCategory=categorizeClassData(dataObj) 
    classMeanDevDiff={} 
    for className,classData in dataCategory.items(): 
        classMeanDevDiff[className] = prepareMeanDevList(classData) 
        print("Data categorization of mean and standard deviation on the basis of className --- ",classMeanDevDiff) 
    return classMeanDevDiff 
     
''' method to calculate probability according to the gaussian formula''' 
def calGaussianFormula(x,mean,deviation): 
    expNumerate = math.pow(x-mean,2) 
    expDenome= 2*math.pow(deviation,2) 
    print(expNumerate,expDenome) 
    exponent = math.exp(-(expNumerate/expDenome)) 
    print("Exponent value according to gaussian formula --- ",exponent) 
    probab= (1/(math.sqrt(2*math.pi)*deviation))*exponent 
    print("Final probability of particular input row according to gaussian formula --- ",probab) 
    return probab 

''' method to calculate normal distribution probability''' 
def normPDF(x, mean, deviation): 
    variance = math.pow(deviation,2) 
    denom = (2*math.pi*variance)**.5 
    num = math.exp(-(float(x)-float(mean))**2/(2*variance)) 
    print("Normal distribution of individual row --- ",num/denom) 
    return num/denom 
  
''' method to calculate the class probabilities''' 
def getClassProbability(classMeanDevDiff,inputRow): 
    classProbabilities={} 
    for className,classMeanDev in classMeanDevDiff.items(): 
        classProbabilities[className] = 1 
        for j in range(len(classMeanDev)): 
            mean,deviation=classMeanDev[j] 
            x=inputRow[j] 
            '''classProbabilities[className]=calGaussianFormula(x,mean,deviation)''' 
            classProbabilities[className]=normPDF(x, mean, deviation) 
    print("Overall class probabilities are --- ",classProbabilities) 
    return classProbabilities 
  
''' method to set the prior class at the identification on the basis of highest probability of class ''' 
def priorClassIdentification(classMeanDevDiff,inputRow): 
    classProbabilities = getClassProbability(classMeanDevDiff,inputRow) 
    selectedClass,maxProbab=None,-1 
    for className,classProbability in classProbabilities.items(): 
        if selectedClass is None or classProbability > maxProbab: 
            maxProbab=classProbability 
            selectedClass=className 
    print("Prior prediction class name and it's probability according to highest probability --- ",selectedClass," --- ",maxProbab) 
    return selectedClass 
     
''' method to check prior class identification on the test data prepared earlier''' 
def checkPriorClass(classMeanDevDiff,testData): 
    predictArr=[] 
    for i in range(len(testData)): 
        priorClass=priorClassIdentification(classMeanDevDiff,testData[i]) 
        predictArr.append(priorClass) 
    print("Cross check prior identification method on some test data sets --- ",predictArr) 
    return predictArr 

''' method to check the accuracy of the prior prediction of test data set ''' 
def checkAccuracy(testData,predictArr): 
    count=0 
    for j in range(len(testData)): 
        tDataClass = testData[j][len(testData[j]) -1]
        if tDataClass == predictArr[j]: 
            count+=1 
    accuracyPercent = (count/float(len(testData)))*100 
    print("Accuracy percent of the test data set --- ",accuracyPercent) 
    return accuracyPercent 

''' method to calculate error percentage from accuracy'''
def getErrorPercent(accuracyPercent):
    return (100.0 - accuracyPercent);

''' method to get actual class array from testdata'''
def getActualClasses(testData):
    actualClass=[]
    for j in range(len(testData)):
        actualClass.append(testData[j][-1])
    return actualClass

''' method to get confusion matrix on the basis of actual class data and prediction data'''
def getConfusionMatrix(actualArr,predictArr):
    actualSeries = pd.Series(actualArr, name='Actual')
    predictSeries = pd.Series(predictArr, name='Predicted')
    confusionMatrix = pd.crosstab(actualSeries, predictSeries, rownames=['Actual'], colnames=['Predicted'], margins=True)
    print(confusionMatrix)
    return np.matrix(confusionMatrix)

'''method to get accuracy using confusion matrix   (TP+TN)/(TP+FP+TN+FN)  '''
def getAccuracyConfusion(confMatrix):
    numer =confMatrix.item(0,0)+confMatrix.item(1,1)
    confAccu=numer/confMatrix.item(2,2)
    return confAccu

'''method to get error using confusion matrix   (FP+FN)/(TP+FP+TN+FN)  '''
def getErrorConfusion(confMatrix):
    numer =confMatrix.item(0,1)+confMatrix.item(1,0)
    confErr=numer/confMatrix.item(2,2)
    return confErr

'''method to calculate sensitivity    TP / (FN + TP)  '''
def getSensitivity(confMatrix):
    return confMatrix.item(1,1)/confMatrix.item(1,2)

'''method to calculate specificity    TN / (TN/FP)  '''
def getSpecificity(confMatrix):
    return confMatrix.item(0,0)/confMatrix.item(0,2)

''' main method to call other methods'''
def runClassifier(): 
    fileData = [["3.393533211","2.331273381","0"], 
        ["3.110073483","1.781539638","0"], 
        ["1.343808831","3.368360954","0"], 
        ["3.582294042","4.67917911","0"], 
        ["2.280362439","2.866990263","0"], 
        ["7.423436942","4.696522875","1"], 
        ["5.745051997","3.533989803","1"], 
        ["1.343808831","3.368360954","0"], 
        ["3.582294042","4.67917911","0"], 
        ["2.280362439","2.866990263","0"], 
        ["7.423436942","4.696522875","1"], 
        ["5.745051997","3.533989803","1"], 
        ["9.172168622","2.511101045","1"], 
        ["7.792783481","3.424088941","1"], 
        ["7.939820817","0.791637231","1"]] 

    ratio=0.65 
    fileData=convertDatatoFloat(fileData) 
    tData,testData=prepareTrainingData(fileData,ratio) 
    classMeanDevDiff = makeClassListMeanDev(tData) 
    testPriorClasses=checkPriorClass(classMeanDevDiff,testData) 
    testAccuracy = checkAccuracy(testData,testPriorClasses) 
    print("Final Accuracy on test data--- ",testAccuracy) 
    predictArr=checkPriorClass(classMeanDevDiff,tData) 
    accuracy = checkAccuracy(tData,predictArr) 
    print("Final Accuracy on training data--- ",accuracy) 
    errors = getErrorPercent(accuracy)
    print("Error percent --- ",errors)
    confusionMatrix = getConfusionMatrix(getActualClasses(tData),predictArr);
    confAcc=getAccuracyConfusion(confusionMatrix)
    print(" Classification Error --- ",confAcc)
    confErr=getErrorConfusion(confusionMatrix)
    print(" Classifier Error --- ",confErr)
    confSense=getSensitivity(confusionMatrix)
    print(" Classifier Sensitivity --- ",confSense)
    confSpec=getSpecificity(confusionMatrix)
    print(" Classifier Specificity --- ",confSpec)

runClassifier(); 
```
